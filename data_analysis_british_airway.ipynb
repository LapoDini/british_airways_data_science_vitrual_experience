```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"..\data\rating_data.csv", parse_dates = ["Date Flown"])
```

```python
df["Recommended"].value_counts().plot(kind = "bar")
plt.xticks(rotation = 0)
plt.title("Positive vs. Negative Reviews", fontsize = 20)
plt.savefig("yes_vs_no.png", transparent = True)
```

```python
df.info()
```

```python
df["Flight Month"] = df["Date Flown"].dt.month
df["Flight Year"] = df["Date Flown"].dt.year
```

```python
df.head()
```

```python
pd.crosstab(df["Type Of Traveller"], values = df["Recommended"], aggfunc = "count", columns = df["Recommended"]).plot(kind = "bar")

plt.title("Recommendation vs. Type of Traveller", fontsize = 20)
plt.xticks(rotation = 0)
plt.savefig("class.png", transparent = True)
plt.show()
```
